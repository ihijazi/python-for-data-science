# Visualizing The Gender Gap In College Degrees

We'll be working with a dataset on the gender gap across college degrees 1970 and 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

> [Downlaod dataset here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv)

---
#### Import needed modules and load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
```

---
##### Visualizing

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)


stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
                 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(18, 20))

counter1 = 0
counter2 = 0
counter3 = 0

for sp in range(1,19):
    if sp == 17:
        continue
    
    ax = fig.add_subplot(6,3,sp)
    
    if sp in [1,4,7,10,13,16]:    
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[counter1]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[counter1]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[counter1])
        counter1+=1
    elif sp in [2,5,8,11,14]:
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[counter2]], c=cb_dark_blue, label='Women', 
                linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[counter2]], c=cb_orange, label='Men', 
                linewidth=3)
        ax.set_title(lib_arts_cats[counter2])
        counter2+=1
    elif sp in [3,6,9,12,15,18]:
        ax.plot(women_degrees['Year'], women_degrees[other_cats[counter3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[counter3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[counter3])
        counter3+=1      
    
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 16:
        ax.text(2005, 60, 'Men')
        ax.text(2001, 35, 'Women')         
    elif sp == 2:
        ax.text(2005, 75, 'Men')
        ax.text(2001, 20, 'Women')  
    elif sp == 3:
        ax.text(2005, 90, 'Men')
        ax.text(2001, 20, 'Women')
    elif sp == 18:
        ax.text(2005, 60, 'Men')
        ax.text(2001, 30, 'Women')
        
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp in [16,14,18]:
        ax.tick_params(labelbottom="on")
        
plt.show()
```
