# Star Wars Analytics

In this notebook, we're going to clean up Star Wars survey and then do analsys on it. To download the dataset used, [click here](StarWars.csv)

#### Importing Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

starwars = pd.read_csv("StarWars.csv",encoding="ISO-8859-1")

starwars.head(3)
```

#### Remove NaN RespondentID

```python
starwars = starwars[pd.notnull(starwars["RespondentID"])]

starwars.head(2)
```

#### Cleaning Up and Cast Yes/No Columns to Boolean

- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

```python
yes_no = {"Yes":True,"No":False}

starwars["Have you seen any of the 6 films in the Star Wars franchise?"] = starwars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
starwars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = starwars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

starwars.head(2)
```

#### Cleaning up and Cast Checkbox Columns to Boolean

```python
mapper = {"Star Wars: Episode I  The Phantom Menace" : True,
          "Star Wars: Episode II  Attack of the Clones" : True,
          "Star Wars: Episode III  Revenge of the Sith" : True,
          "Star Wars: Episode IV  A New Hope" : True,
          "Star Wars: Episode V The Empire Strikes Back" : True,
          "Star Wars: Episode VI Return of the Jedi" : True,
          np.nan : False}

counter = 1

for s in starwars.columns[3:9]:
    starwars[s] = starwars[s].map(mapper)
    starwars.rename(columns={s:"seen_" + str(counter)},inplace=True)
    counter+=1
    
starwars.head(2)
```

#### Cleaning Up Ranking Columns: Coverting Data Type to Flloat and Rename Column Names
* 1 means most favorite
* 6 means least favorite

```python
starwars[starwars.columns[9:15]] = starwars[starwars.columns[9:15]].astype(float)
    
starwars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"rank1",
                         "Unnamed: 10" : "rank2","Unnamed: 11":"rank3","Unnamed: 12":"rank4","Unnamed: 13":"rank5",
                         "Unnamed: 14":"rank6"},inplace=True)

counter = 1

for s in starwars.columns[9:15]:
    starwars.rename(columns={s:"ranking_" + str(counter)},inplace=True)
    counter+=1
    
starwars[starwars.columns[9:15]].head(3)
```

#### The Highest Ranked Movie

```python
import seaborn as sns

sns.barplot(data=starwars[starwars.columns[9:15]])
sns.plt.show()
```

#### The Most Seen Movie

```python
sns.barplot(data=starwars[starwars.columns[3:9]])
sns.plt.show()
```

#### Exploring the Data by Binary Segments: Gender

```python
males = starwars[starwars["Gender"] == "Male"]
females = starwars[starwars["Gender"] == "Female"]

fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1 = sns.barplot(data=males[males.columns[3:9]])
ax1.set_title("Males")


ax2 = fig.add_subplot(122)
ax2 = sns.barplot(data=females[females.columns[3:9]])
ax2.set_title("Females")

plt.show()
```

#### Exploring the Data by Non-Binary Segments: Education

```python
fig = plt.figure(figsize=(20,5))

counter = 1

for e in set(starwars["Education"]):
    if pd.isnull(e):
        continue
    edu = starwars[starwars["Education"] == e][starwars.columns[3:9]]
    ax = fig.add_subplot(1,6,counter)
    sn = sns.barplot(data=edu)
    sn.set_title(e)
    plt.xticks(rotation=90)
    counter+=1


plt.show()
```

```python
fig = plt.figure(figsize=(20,8))

counter = 1

for l in set(starwars["Location (Census Region)"]):
    if pd.isnull(l):
        continue
    loc = starwars[starwars["Location (Census Region)"] == l][starwars.columns[3:9]]
    ax = fig.add_subplot(2,5,counter)
    sn = sns.barplot(data=loc)
    sn.set_title(l)
    plt.xticks(rotation=90)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    counter+=1

plt.show()
```

#### Which Character is the Most Liked?
#### Which Character is the Most Disliked?
#### Which Character Creates the Most Controversy?

```python
mapper = {np.nan: -1,
          "Very favorably":5,
          "Somewhat favorably":4,
          "Neither favorably nor unfavorably (neutral)":3,
          "Somewhat unfavorably": 2,
          "Very unfavorably": 1,
          "Unfamiliar":0,
          }

heros = ["Han Solo","Luke Skywalker","Princess Leia Organa","Anakin Skywalker","Obi Wan Kenobi","Emperor Palpatine","Darth Vader","Lando Calrissian","Boba Fett","C-3P0","R2 D2","Jar Jar Binks","Padme Amidala","Yoda"]

i = 0

for c in starwars[starwars.columns[15:29]]:
    starwars[c].map(mapper)
    starwars[c] = starwars[c][starwars[c] != -1]
    starwars.rename(columns={c:heros[i]},inplace=True)
    i=i+1
    
rating_labels = ["Very favorably","Somewhat favorably","Neutral","Somewhat Unfavorably","Very unfavorably","Unfamiliar"]


fig = plt.figure(figsize=(20,30))

counter = 1

xvals = [0,1,2,3,4,5]


for h in heros:
    ax = fig.add_subplot(5,3,counter)
    mysns = sns.barplot(x=xvals,y=starwars[h].value_counts(),ax=ax)
    mysns.set_title(h)
    mysns.set_xticklabels(rating_labels)
    plt.xticks(rotation=30)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.ylim(0,650)
    
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x()+0.2 * 1.005, p.get_height() * 1.005))
    
    counter+=1
    

sns.plt.show()
```
